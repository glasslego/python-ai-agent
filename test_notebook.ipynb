```python
from nltk.sentiment import SentimentIntensityAnalyzer

# 감정 분석기 초기화
analyzer = SentimentIntensityAnalyzer()

# 텍스트 분석
text = "It's nice day!"
result = analyzer.polarity_scores(text)

print(f"긍정: {result['pos']}")
print(f"부정: {result['neg']}")
print(f"중성: {result['neu']}")
print(f"종합: {result['compound']}")

# 최종 감정 판단
if result["compound"] >= 0.05:
    print("😊 긍정적인 감정")
elif result["compound"] <= -0.05:
    print("😞 부정적인 감정")
else:
    print("😐 중성적인 감정")
```
